```python
import pandas as pd
import mysql.connector
import json

# 1) อ่านไฟล์ JSON
df = pd.read_json(
    r"C:\Users\Lenovo\OneDrive\เอกสาร\ProjectDGT\Project 1101910\DataEngineer_JOBFIT\JobGov1\Result\Final_JobGov1.json",
    lines=True
)

# 2) เชื่อม MySQL
conn = mysql.connector.connect(
    host="localhost",
    user="root",
    password="Tenten47rainnet.",
    database="jobplatform",
    port=3306
)
cursor = conn.cursor()

# 3) insert
sql = """
INSERT INTO jobs
(province, position, organization, degree, major, skill, certification, job_type)
VALUES (%s,%s,%s,%s,%s,%s,%s,%s)
"""

for _, row in df.iterrows():
    cursor.execute(sql, (
        row["province"],
        row["position"],
        row["organization"],
        json.dumps(row["degree"]) if row["degree"] else None,
        json.dumps(row["major"]) if row["major"] else None,
        json.dumps(row["skill"]) if row["skill"] else None,
        json.dumps(row["certification"]) if row["certification"] else None,
        row["type"]
    ))

conn.commit()
cursor.close()
conn.close()

print("✅ อัปโหลดสำเร็จ:", len(df), "แถว")
```
